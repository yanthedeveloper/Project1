```python
#Define the list for the desired zip codes
zipCodes = ["30314", "30310", "30312", "30308", "30306", "30311", "30318", "30307", "30309"]

#initialize variables to store data for the desired attributes
family_size = []
num_households = []
crime_score = []

#for loop to hit each zip code and pull data
for code in zipCodes:
    
    conn = http.client.HTTPSConnection("api.gateway.attomdata.com")
     
    headers = { 
    'accept': "application/json", 
    'apikey': "71a48c8fc563111ab89c14d39ebafcdc", 
    }
    
    conn.request("GET", f"/communityapi/v2.0.0/area/full?AreaID=ZI{code}" , headers=headers) 
    
    res = conn.getresponse()
    info = res.read()
    data = json.loads(info)
    
    family_size.append(data["response"]["result"]["package"]["item"][0]["famavesz"])
    num_households.append(data["response"]["result"]["package"]["item"][0]["hhdfam"])
    crime_score.append(data["response"]["result"]["package"]["item"][0]["cocrmcytotc"])
    
print(family_size)
print(num_households)
print(crime_score)
```

```python
df = pd.DataFrame({"Zip Cods": zipCodes, "Family Size": family_size, 
                  "Num. Family Households": num_households, "Crime Score":
                  crime_score})
df
```
